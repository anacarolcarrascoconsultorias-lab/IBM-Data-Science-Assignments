```python
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMSkillsNetwork-AI0272EN-SkillsNetwork/labs/dataset/2016.csv"
```

```python
import pandas as pd

df = pd.read_csv(URL)
display(df.head())
```

```python
df.info()
```

```python
import numpy as np

# Get object columns (likely strings)
object_cols = df.select_dtypes(include='object').columns

for col in object_cols:
    # Remove leading and trailing whitespaces
    df[col] = df[col].astype(str).str.strip()
    # Replace empty strings with NaN
    df[col] = df[col].replace('', np.nan)

# Convert columns to appropriate data types using pandas' best guess
df = df.convert_dtypes()

print("DataFrame after cleaning and type conversion:")
df.info()
```

```python
print("Missing values before imputation:")
print(df.isnull().sum())

# Identify columns with missing values
missing_cols = df.columns[df.isnull().any()].tolist()

# Impute missing values with the mean of each column
for col in missing_cols:
    if pd.api.types.is_numeric_dtype(df[col]):
        col_mean = df[col].mean()
        df[col] = df[col].fillna(col_mean)
        print(f"Filled missing values in column '{col}' with mean: {col_mean:.2f}")
    else:
        print(f"Column '{col}' is not numeric; skipping mean imputation.")

print("\nMissing values after imputation:")
print(df.isnull().sum())

print("\nDataFrame info after imputation:")
df.info()
```

```python
print("\n--- Number of rows for each Country ---")
country_counts = df['Country'].value_counts()
print(country_counts)

print("\n--- Average Happiness Score per Region ---")
avg_happiness_by_region = df.groupby('Region')['Happiness Score'].mean()
print(avg_happiness_by_region)
```

```python
#Task 1: Identify the GDP per capita and Healthy Life Expectancy of the top 10 countries.
## Write a python code that identifies the GDP per capita and Healthy Life Expectancy of the top 10 countries and create a bar chart named fig1 to show the GDP per capita and Healthy Life Expectancy of these top 10 countries using plotly.

import plotly.graph_objects as go

# Ensure the columns for plotting are numeric. From previous df.info(), they were strings.
# Need to convert them to numeric, handling potential errors.

# Convert 'Economy (GDP per Capita)' to numeric
# Some values might be non-numeric, so use errors='coerce' to turn them into NaN
df['Economy (GDP per Capita)'] = pd.to_numeric(df['Economy (GDP per Capita)'], errors='coerce')

# Convert 'Health (Life Expectancy)' to numeric
df['Health (Life Expectancy)'] = pd.to_numeric(df['Health (Life Expectancy)'], errors='coerce')

# Impute any newly created NaNs after conversion with the mean of the column
# This step ensures that all values are numeric for plotting
df['Economy (GDP per Capita)'] = df['Economy (GDP per Capita)'].fillna(df['Economy (GDP per Capita)'].mean())
df['Health (Life Expectancy)'] = df['Health (Life Expectancy)'].fillna(df['Health (Life Expectancy)'].mean())

# Get the top 10 countries by Happiness Score
top_10_countries = df.nlargest(10, 'Happiness Score')

# Create a bar chart using Plotly
fig1 = go.Figure()

fig1.add_trace(go.Bar(
    x=top_10_countries['Country'],
    y=top_10_countries['Economy (GDP per Capita)'],
    name='Economy (GDP per Capita)',
    marker_color='indianred'
))

fig1.add_trace(go.Bar(
    x=top_10_countries['Country'],
    y=top_10_countries['Health (Life Expectancy)'],
    name='Health (Life Expectancy)',
    marker_color='lightsalmon'
))

fig1.update_layout(
    title='Top 10 Happiest Countries: GDP per Capita and Healthy Life Expectancy',
    xaxis_title='Country',
    yaxis_title='Value',
    barmode='group'
)

fig1.show()
```

```python
# Task 2: Find the correlation between the Economy (GDP per Capita), Family, Health (Life Expectancy), Freedom, Trust (Government Corruption), Generosity and Happiness score. You may like to represent the correlation as a heatmap of a readable, visually appealing size.
## Write a python code that performs the following actions: 1. Create a sub-dataset including Economy (GDP per Capita), Family, Health (Life Expectancy), Freedom, Trust (Government Corruption), Generosity, and Happiness Score attributes from the dataframe (df); 2. Find the correlation between the attributes in the subdataset as a heatmap named fig2 using Plotly of width 800 and height 600.

import plotly.graph_objects as go
import pandas as pd

# 1. Create a sub-dataset with the specified attributes
correlation_columns = [
    'Economy (GDP per Capita)',
    'Family',
    'Health (Life Expectancy)',
    'Freedom',
    'Trust (Government Corruption)',
    'Generosity',
    'Happiness Score'
]

# Ensure all specified columns are numeric before calculating correlation
# Convert 'Economy (GDP per Capita)' to numeric, coercing errors to NaN
df['Economy (GDP per Capita)'] = pd.to_numeric(df['Economy (GDP per Capita)'], errors='coerce')
# Convert 'Health (Life Expectancy)' to numeric, coercing errors to NaN
df['Health (Life Expectancy)'] = pd.to_numeric(df['Health (Life Expectancy)'], errors='coerce')
# Impute any NaNs in these columns with their mean to ensure correlation can be calculated
df['Economy (GDP per Capita)'] = df['Economy (GDP per Capita)'].fillna(df['Economy (GDP per Capita)'].mean())
df['Health (Life Expectancy)'] = df['Health (Life Expectancy)'].fillna(df['Health (Life Expectancy)'].mean())

sub_df = df[correlation_columns]

# 2. Find the correlation between the attributes
correlation_matrix = sub_df.corr()

# Create a heatmap using Plotly
fig2 = go.Figure(data=go.Heatmap(
                   z=correlation_matrix.values,
                   x=correlation_matrix.columns,
                   y=correlation_matrix.index,
                   colorscale='RdBu',
                   colorbar=dict(title='Correlation Coefficient')))

fig2.update_layout(
    title='Correlation Matrix of Happiness Factors',
    xaxis_title='Factors',
    yaxis_title='Factors',
    width=800,
    height=600,
    xaxis_showgrid=False,
    yaxis_showgrid=False,
    yaxis_autorange='reversed'
)

fig2.show()
```

```python
# Task 3:Create a scatter plot to identify the effect of GDP per Capita on Happiness Score in various Regions. Use plotly for creating the plot.
## Write a code that creates a scatter plot named fig3 between Happiness Score and GDP per Capita attributes of a dataframe using Plotly. Use Region to color the data points on the scatter plot.

import plotly.express as px

# Create a scatter plot
fig3 = px.scatter(df,
                  x='Economy (GDP per Capita)',
                  y='Happiness Score',
                  color='Region',
                  title='Happiness Score vs. GDP per Capita by Region')

fig3.update_layout(
    xaxis_title='Economy (GDP per Capita)',
    yaxis_title='Happiness Score'
)

fig3.show()
```

```python
# Task 4: Create a pie chart to present Happiness Score by Regions
## Write a Plotly code that creates a pie chart named fig4 to present Happiness Score by Region attributes of dataframe df.

import plotly.express as px

# Calculate the average Happiness Score for each Region
avg_happiness_by_region = df.groupby('Region')['Happiness Score'].mean().reset_index()

# Create a pie chart
fig4 = px.pie(avg_happiness_by_region,
              values='Happiness Score',
              names='Region',
              title='Average Happiness Score by Region',
              hole=0.3) # Optional: creates a donut chart

fig4.show()
```

```python
# Task 5: Create a map to display GDP per capita of countries and include Healthy life expectancy to be shown as a tooltip
## Write a Plotly code that creates a map named fig5 to display GDP per capita of countries and include Healthy Life Expectancy to be shown as a tooltip.

import plotly.express as px

# Create a choropleth map
fig5 = px.choropleth(df,
                    locations='Country',
                    locationmode='country names',
                    color='Economy (GDP per Capita)',
                    hover_name='Country',
                    hover_data=['Health (Life Expectancy)'],
                    color_continuous_scale=px.colors.sequential.Plasma,
                    title='GDP per Capita by Country with Healthy Life Expectancy Tooltip')

fig5.show()
```

```python
# Task 6: Write prompts that generate codes to write at least four of the graph plots generated in the previous steps into a HTML page.
## Write Python code to write any four of the Plotly figures (fig1, fig2, fig3, fig4, fig5) to a single HTML file named dashboard.html

# Write selected Plotly figures to a single HTML file

# Ensure figures are defined. If this cell is run independently,
# make sure fig1, fig2, fig3, fig4 are available in the environment.
# In this notebook, they have been defined in previous executed cells.

# Generate HTML for each figure
fig1_html = fig1.to_html(full_html=False, include_plotlyjs='cdn')
fig2_html = fig2.to_html(full_html=False, include_plotlyjs='cdn')
fig3_html = fig3.to_html(full_html=False, include_plotlyjs='cdn')
fig4_html = fig4.to_html(full_html=False, include_plotlyjs='cdn')

# Combine the HTML components into a single dashboard HTML structure
dashboard_html_content = f"""
<!DOCTYPE html>
<html>
<head>
    <title>Plotly Dashboard</title>
    <script src="https://cdn.plot.ly/plotly-latest.min.js"></script>
</head>
<body>
    <h1>Plotly Figures Dashboard</h1>
    <h2>Figure 1: Top 10 Happiest Countries: GDP per Capita and Healthy Life Expectancy</h2>
    {fig1_html}
    <h2>Figure 2: Correlation Matrix of Happiness Factors</h2>
    {fig2_html}
    <h2>Figure 3: Happiness Score vs. GDP per Capita by Region</h2>
    {fig3_html}
    <h2>Figure 4: Average Happiness Score by Region</h2>
    {fig4_html}
</body>
</html>
"""

# Write the combined HTML content to a file
file_name = "dashboard.html"
with open(file_name, 'w') as f:
    f.write(dashboard_html_content)

print(f"Dashboard saved to {file_name}")
```
