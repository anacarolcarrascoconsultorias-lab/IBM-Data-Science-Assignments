```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect('survey-results-public.sqlite')
```

## Demo: Basic SQL Queries

#### Demo 3: Group Data by Age


```python
QUERY = """
SELECT Age, COUNT(*) as count
FROM main
GROUP BY Age
ORDER BY Age
"""
df_age = pd.read_sql_query(QUERY, conn)
print(df_age)
```

```python
## INSPECTING VALUES INSIDE JOBSATPOINTS_6
# Inspect distinct values
query = """
SELECT JobSatPoints_6
FROM main
"""
df_tmp = pd.read_sql_query(query, conn)

# Look at unique values (limited to first 30 so it doesn't explode)
df_tmp['JobSatPoints_6'].dropna().unique()[:30]
```

```python
df_tmp['JobSatPoints_6'].value_counts(dropna=False).head(20)
```

```python
## CLEANING THE COLUMNS USED IN THE LAB]
# --------------------------------------
# CLEANING: CompTotal, Age, JobSatPoints_6
# --------------------------------------

# Load the full table
df = pd.read_sql_query("SELECT * FROM main", conn)

# ========== CLEAN CompTotal ==========
df['CompTotal'] = pd.to_numeric(df['CompTotal'], errors='coerce')

# Remove negative or zero compensation
df = df[df['CompTotal'] > 0]

# Remove extreme outliers (same rule used before)
df = df[df['CompTotal'] < 1_000_000]


# ========== CLEAN Age ==========

age_map = {
    "Under 18 years old": 15,
    "18-24 years old": 21,
    "25-34 years old": 29.5,
    "35-44 years old": 39.5,
    "45-54 years old": 49.5,
    "55-64 years old": 59.5,
    "65 years old or older": 70
}

df['Age_numeric'] = df['Age'].map(age_map)


# ========== CLEAN JobSatPoints_6 ==========
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Keep only values in a valid range (0–100)
df.loc[(df['JobSatPoints_6'] < 0) | (df['JobSatPoints_6'] > 100), 'JobSatPoints_6'] = None
```

## Visualizing Data


### Task 1: Visualizing the Distribution of Data


**1. Box Plot of `CompTotal` (Total Compensation)**


Use a box plot to analyze the distribution and outliers in total compensation.


```python
# your code goes here
plt.figure(figsize=(8,5))
plt.boxplot(df['CompTotal'].dropna(), vert=True)
plt.title("Box Plot of CompTotal (Cleaned)")
plt.ylabel("Total Compensation")
plt.show()
```

**2. Box Plot of Age (converted to numeric values)**


Convert the `Age` column into numerical values and visualize the distribution.


```python
# your code goes here
plt.figure(figsize=(8,5))
plt.boxplot(df['Age_numeric'].dropna(), vert=True)
plt.title("Box Plot of Age (Mapped to Numeric Values)")
plt.ylabel("Age (Numeric)")
plt.show()
```

```python
df['Age'].value_counts(dropna=False).head(20)
```

### Task 2: Visualizing Relationships in Data


**1. Box Plot of `CompTotal` Grouped by Age Groups:**


Visualize the distribution of compensation across different age groups.


```python
# your code goes here
# Create readable age bins based on Age_numeric
bins = [0, 24, 34, 44, 54, 64, 120]
labels = ["18-24", "25-34", "35-44", "45-54", "55-64", "65+"]

df['Age_group'] = pd.cut(df['Age_numeric'], bins=bins, labels=labels)
```

```python
plt.figure(figsize=(10,6))
df.boxplot(column='CompTotal', by='Age_group', grid=False)
plt.title("CompTotal Distribution by Age Group")
plt.suptitle("")  # removes unwanted default title
plt.xlabel("Age Group")
plt.ylabel("Total Compensation")
plt.xticks(rotation=45)
plt.show()
```

**2. Box Plot of `CompTotal` Grouped by Job Satisfaction (`JobSatPoints_6`):**


Examine how compensation varies based on job satisfaction levels.


```python
# your code goes here
df_plot = df.dropna(subset=['CompTotal', 'JobSatPoints_6'])

plt.figure(figsize=(12,6))
df_plot.boxplot(column='CompTotal', by='JobSatPoints_6', grid=False)
plt.title("CompTotal Distribution by Job Satisfaction Score")
plt.suptitle("")  # remove extra matplotlib title
plt.xlabel("Job Satisfaction Score")
plt.ylabel("Total Compensation")
plt.xticks(rotation=90)
plt.show()
```

```python
bins = [0, 20, 45, 65, 85, 100]
labels = ["Very Low", "Low", "Medium", "High", "Very High"]

df['JobSat_group'] = pd.cut(df['JobSatPoints_6'], bins=bins, labels=labels)
```

```python
plt.figure(figsize=(10,6))
df.dropna(subset=['CompTotal','JobSat_group']) \
  .boxplot(column='CompTotal', by='JobSat_group', grid=False)

plt.title("CompTotal by Job Satisfaction Group")
plt.suptitle("")
plt.xlabel("Job Satisfaction Group")
plt.ylabel("Total Compensation")
plt.show()
```

### Task 3: Visualizing the Composition of Data


**1. Box Plot of `ConvertedCompYearly` for the Top 5 Developer Types:**


Analyze compensation across the top 5 developer roles.


```python
# your code goes here
df.columns
```

```python
# --- Task 3.1: ConvertedCompYearly by Top 5 Developer Types ---

# Make sure compensation is numeric
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')

# Keep only rows where both DevType and ConvertedCompYearly are present
df_dev = df.dropna(subset=['DevType', 'ConvertedCompYearly']).copy()

# Split multi-select DevType string into a list
df_dev['DevType_list'] = df_dev['DevType'].str.split(';')

# Explode so each developer type becomes its own row
df_dev = df_dev.explode('DevType_list')

# Strip extra spaces from each type
df_dev['DevType_clean'] = df_dev['DevType_list'].str.strip()

# Drop any empty strings just in case
df_dev = df_dev[df_dev['DevType_clean'] != '']
```

```python
# Find the top 5 most common developer types
top5_devtypes = df_dev['DevType_clean'].value_counts().head(5).index

top5_devtypes
```

```python
# Filter to only those top 5 developer types
df_top5_dev = df_dev[df_dev['DevType_clean'].isin(top5_devtypes)]

# Create the box plot
plt.figure(figsize=(12,6))
df_top5_dev.boxplot(column='ConvertedCompYearly',
                    by='DevType_clean',
                    grid=False)

plt.title("ConvertedCompYearly for Top 5 Developer Types")
plt.suptitle("")  # remove automatic extra title
plt.xlabel("Developer Type")
plt.ylabel("Yearly Compensation")
plt.xticks(rotation=45)
plt.show()
```

```python
import textwrap

# Wrap long x-axis labels for readability
df_top5_dev['DevType_wrapped'] = df_top5_dev['DevType_clean'].apply(
    lambda x: "\n".join(textwrap.wrap(x, width=20))
)

plt.figure(figsize=(14,7))

df_top5_dev.boxplot(
    column='ConvertedCompYearly',
    by='DevType_wrapped',
    grid=False
)

plt.title("ConvertedCompYearly for Top 5 Developer Types")
plt.suptitle("")   # Remove extra matplotlib title
plt.xlabel("Developer Type")
plt.ylabel("Yearly Compensation")

# Adjust layout so labels fit
plt.xticks(rotation=45)   # No rotation needed after wrapping
plt.tight_layout()

plt.show()
```

```python
import textwrap

# Wrap long DevType labels
df_top5_dev['DevType_wrapped'] = df_top5_dev['DevType_clean'].apply(
    lambda x: "\n".join(textwrap.wrap(x, width=15))
)

plt.figure(figsize=(14,7))

df_top5_dev.boxplot(
    column='ConvertedCompYearly',
    by='DevType_wrapped',
    grid=False
)

plt.title("ConvertedCompYearly for Top 5 Developer Types")
plt.suptitle("")  # remove automatic title
plt.xlabel("Developer Type")
plt.ylabel("Yearly Compensation")
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

**2. Box Plot of `CompTotal` for the Top 5 Countries:**


Analyze compensation across respondents from the top 5 countries.


```python
# your code goes here
df_country = df.dropna(subset=['Country', 'CompTotal']).copy()
df_country['CompTotal'] = pd.to_numeric(df_country['CompTotal'], errors='coerce')
df_country = df_country[df_country['CompTotal'] > 0]      # keep valid comp
```

```python
import textwrap

# Wrap long country names into multiple lines
df_top5_country['Country_wrapped'] = df_top5_country['Country'].apply(
    lambda x: "\n".join(textwrap.wrap(x, width=12))
)

plt.figure(figsize=(15,25))

df_top5_country.boxplot(
    column='CompTotal',
    by='Country_wrapped',
    grid=False
)

plt.title("CompTotal for Top 5 Countries")
plt.suptitle("")   # remove default title
plt.xlabel("Country")
plt.ylabel("Total Compensation")

plt.xticks(rotation=0)   # not needed anymore because labels are wrapped
plt.tight_layout()
plt.show()
```

### Task 4: Visualizing Comparison of Data


**1. Box Plot of CompTotal Across Employment Types:**


Analyze compensation for different employment types.


```python
# your code goes here
df_emp = df.dropna(subset=['Employment', 'CompTotal']).copy()
df_emp['CompTotal'] = pd.to_numeric(df_emp['CompTotal'], errors='coerce')
df_emp = df_emp[df_emp['CompTotal'] > 0]
```

```python
# Start from rows where we have Employment and CompTotal
df_emp_raw = df.dropna(subset=['Employment', 'CompTotal']).copy()

# Make sure compensation is numeric and positive
df_emp_raw['CompTotal'] = pd.to_numeric(df_emp_raw['CompTotal'], errors='coerce')
df_emp_raw = df_emp_raw[df_emp_raw['CompTotal'] > 0]

# Split multi-select Employment into list
df_emp_raw['Employment_list'] = df_emp_raw['Employment'].str.split(';')

# Explode into one row per employment type
df_emp = df_emp_raw.explode('Employment_list').copy()

# Strip spaces and drop empties
df_emp['Employment_clean'] = df_emp['Employment_list'].str.strip()
df_emp = df_emp[df_emp['Employment_clean'] != '']

```

```python
df_emp['Employment_clean'].value_counts().head(20)
```

```python
top5_emp = df_emp['Employment_clean'].value_counts().head(5).index
top5_emp
```

```python
import textwrap

df_emp_top = df_emp[df_emp['Employment_clean'].isin(top5_emp)].copy()

# Wrap long labels onto multiple lines
df_emp_top['Employment_wrapped'] = df_emp_top['Employment_clean'].apply(
    lambda x: "\n".join(textwrap.wrap(x, width=15))
)

plt.figure(figsize=(16,25))
df_emp_top.boxplot(
    column='CompTotal',
    by='Employment_wrapped',
    grid=False
)

plt.title("CompTotal Across Top 5 Employment Types")
plt.suptitle("")            # remove extra default title
plt.xlabel("Employment Type")
plt.ylabel("Total Compensation")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

**2. Box Plot of `YearsCodePro` by Job Satisfaction (`JobSatPoints_6`):**


Examine the distribution of professional coding years by job satisfaction levels.


```python
# your code goes here
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')
df_ycp = df.dropna(subset=['YearsCodePro', 'JobSatPoints_6']).copy()

plt.figure(figsize=(14,8))

df_ycp.boxplot(
    column='YearsCodePro',
    by='JobSatPoints_6',
    grid=False
)

plt.title("Years of Professional Coding Experience by Job Satisfaction Score")
plt.suptitle("")  # remove default matplotlib title
plt.xlabel("Job Satisfaction Score")
plt.ylabel("Years of Professional Coding Experience")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# PERSONAL INSIGHTS
##Summary & Interpretation — Box Plots

In this lab, I explored several aspects of the Stack Overflow survey dataset using box plots, focusing on distributions, relationships, and category-based comparisons. Before generating any visualizations, essential data-cleaning steps were performed to ensure accuracy and interpretability. These included converting compensation and satisfaction values to numeric types, mapping age brackets to numerical midpoints, cleaning multi-select fields (Developer Type and Employment), splitting and exploding lists into individual rows, and removing invalid or extreme values.

**📌 1. Distribution Analysis
CompTotal (Total Compensation)**

A box plot was generated after cleaning negative and unrealistic values (> 1,000,000).
Interpretation:
The compensation distribution is right-skewed, with a large number of high-value outliers. The majority of respondents fall within a moderate compensation range, while a small portion earns significantly more, producing the long upper whisker and dense outlier region.

Age (Mapped to Numeric Values)

Age categories were mapped to numerical midpoints and plotted.
Interpretation:
Most respondents are between their mid-20s and mid-40s. The distribution is relatively compact, with a few respondents in the oldest category appearing as mild outliers. This aligns with the typical demographic of technology-related surveys.

**📌 2. Relationship Analysis
CompTotal by Age Group**

Age was binned into categories (18–24, 25–34, 35–44, etc.).
Interpretation:
Compensation tends to increase slightly with age, particularly between the younger (18–24) and mid-career (35–44) groups. However, high variability persists in all age ranges, which is expected in tech compensation because salary depends heavily on role, region, and specialization—not age alone.

**CompTotal by Job Satisfaction Score**

Raw job satisfaction scores produced many categories, so the visualization showed substantial clutter. Though technically accurate, the plot illustrated that no clear linear relationship exists between satisfaction and compensation.
Interpretation:
Higher compensation does not reliably predict higher job satisfaction. Respondents at every satisfaction level show wide compensation ranges, suggesting satisfaction is driven more by environment, workload, culture, and management rather than salary alone.

**📌 3. Composition Analysis
ConvertedCompYearly for Top 5 Developer Types**

Developer Type is a multi-select column, so it was split on semicolons, cleaned, exploded into individual rows, and counted.
Interpretation:
The top 5 developer roles (back-end, front-end, full-stack, desktop/enterprise, embedded/IoT) showed similar compensation distributions, although back-end and full-stack developers exhibited slightly higher medians. All roles displayed significant outliers, reinforcing that specialization, region, and experience strongly influence pay.

**CompTotal for Top 5 Countries**

After identifying the five most represented countries, compensation distributions were compared across them.
Interpretation:
Countries differed noticeably in compensation levels, reflecting cost-of-living differences and regional salary structures. For example, the United States typically showed higher median compensation and a wider spread, while European countries had more compressed distributions.

**📌 4. Comparison Analysis
CompTotal Across Top 5 Employment Types**

Employment Type is also multi-select and was cleaned the same way as Developer Type.
Interpretation:
Full-time workers showed predictably higher compensation medians, while part-time workers and students showed lower ranges. Independent contractors exhibited higher variability—some earning significantly more due to contract-based income, others less depending on role and project load.

**YearsCodePro by Job Satisfaction Group**

Years of professional coding experience was cleaned and plotted against grouped satisfaction scores.
Interpretation:
Respondents with higher job satisfaction tended to have slightly more years of experience on average, but distributions overlap across all groups. This suggests that while experience may contribute to satisfaction (e.g., confidence, autonomy), other factors are equally or more influential.
